```python
# not front end
# take 30 img

import cv2
import numpy as np
import os
from tkinter import *
from tkinter import messagebox
import pandas as pd
from datetime import datetime
import time


# Ensure these folders exist
if not os.path.exists("TrainingImage"):
    os.makedirs("TrainingImage")
if not os.path.exists("TrainingImageLabel"):
    os.makedirs("TrainingImageLabel")
if not os.path.exists("Attendance"):
    os.makedirs("Attendance")

# Path to Haar cascade XML file - update if you place it elsewhere
haarcascadePath = "haarcascade_frontalface_default.xml"
detector = cv2.CascadeClassifier(haarcascadePath)
if detector.empty():
    print("Cascade file not loaded! Check path and file.")

# Function to take images for a student ID and name
def take_images(Id, name, notify):
    if not Id.isdigit() or not name.isalpha():
        notify.set("Invalid ID or Name!")
        return

    detector = cv2.CascadeClassifier(haarcascadePath)
    if detector.empty():
        notify.set("Cascade not loaded!")
        return

    cam = cv2.VideoCapture(0)

    directions = ["Front", "Left", "Right"]   # guide user
    images_needed = 15                        # per direction
    captured = {d: 0 for d in directions}     # tracking count

    current_dir = 0
    font = cv2.FONT_HERSHEY_SIMPLEX

    while current_dir < len(directions):
        ret, img = cam.read()
        if not ret:
            break

        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = detector.detectMultiScale(gray, 1.3, 5)

        # --- Show instruction on Tkinter label too ---
        notify.set(f"Please look {directions[current_dir]} ({captured[directions[current_dir]]}/{images_needed})")
        notify_label = notify._tk.globalgetvar(notify._name)  # hacky check, or call root.update()
        try:
            notify_label  # avoids error
            notify._tk.call("update")  # refresh UI
        except:
            pass

        for (x, y, w, h) in faces:
            captured[directions[current_dir]] += 1
            count = captured[directions[current_dir]]
            cv2.imwrite(f"TrainingImage/{name}.{Id}.{directions[current_dir]}.{count}.jpg", gray[y:y+h, x:x+w])

            cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
            cv2.putText(img, f"{directions[current_dir]}: {count}/{images_needed}", 
                        (10, 450), font, 0.8, (0, 255, 255), 2)

            time.sleep(0.2)

            if count >= images_needed:
                current_dir += 1
                break

        cv2.imshow("Taking Images", img)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cam.release()
    cv2.destroyAllWindows()
    notify.set(f"Images captured for {name} (Front/Left/Right)")


# Function to train the recognizer
def train_images():
    recognizer = cv2.face.LBPHFaceRecognizer_create()
    
    def getImagesAndLabels(path):
        imagePaths = [os.path.join(path, f) for f in os.listdir(path)]
        faces = []
        Ids = []
        for imagePath in imagePaths:
            pil_img = cv2.imread(imagePath, cv2.IMREAD_GRAYSCALE)
            if pil_img is None:
                continue
            Id = int(os.path.split(imagePath)[-1].split(".")[1])
            faces.append(pil_img)
            Ids.append(Id)
        return faces, Ids

    faces, Ids = getImagesAndLabels("TrainingImage")
    if len(faces) == 0:
        messagebox.showerror("Error", "No images found to train. Please take images first.")
        return
    recognizer.train(faces, np.array(Ids))
    recognizer.save("TrainingImageLabel/Trainer.yml")
    messagebox.showinfo("Result", "Training datasets completed!")

# Function to mark attendance in CSV file
def mark_attendance(Id, name):
    date_str = datetime.now().strftime("%Y-%m-%d")
    time_str = datetime.now().strftime("%H:%M:%S")
    filename = f"Attendance/Attendance_{date_str}.csv"
    if not os.path.exists(filename):
        df = pd.DataFrame(columns=["Id", "Name", "Date", "Time"])
    else:
        df = pd.read_csv(filename)

    # Use concat instead of append (pandas 2.0+)
    if not ((df["Id"] == Id) & (df["Date"] == date_str)).any():
        df = pd.concat([df, pd.DataFrame([{"Id": Id, "Name": name, "Date": date_str, "Time": time_str}])], ignore_index=True)
        df.to_csv(filename, index=False)

# Function to recognize faces and mark attendance with immediate camera close after detection
def recognize_faces():
    recognizer = cv2.face.LBPHFaceRecognizer_create()
    if not os.path.exists("TrainingImageLabel/Trainer.yml"):
        messagebox.showerror("Error", "Trainer file not found, please train images first.")
        return
    recognizer.read("TrainingImageLabel/Trainer.yml")

    faceCascade = cv2.CascadeClassifier(haarcascadePath)
    if faceCascade.empty():
        messagebox.showerror("Error", f"Failed to load cascade classifier xml file '{haarcascadePath}'. Please check the file and path.")
        return

    font = cv2.FONT_HERSHEY_SIMPLEX
    cam = cv2.VideoCapture(0)
    detected_face = False  # Flag to break loop after detection
    while True:
        ret, img = cam.read()
        if not ret:
            break
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = faceCascade.detectMultiScale(gray, 1.2, 5)
        for (x, y, w, h) in faces:
            Id, conf = recognizer.predict(gray[y:y+h, x:x+w])
            if conf < 50:
                matching_files = [f for f in os.listdir("TrainingImage") if f.split(".")[1] == str(Id)]
                name = matching_files[0].split(".")[0] if matching_files else "Unknown"
                color = (0, 255, 0)
                mark_attendance(Id, name)
            else:
                name = "Unknown"
                color = (0, 0, 255)

            cv2.putText(img, f"{name}", (x, y-10), font, 0.8, color, 2)
            cv2.rectangle(img, (x, y), (x+w, y+h), color, 3)
            detected_face = True

        cv2.imshow("Face Recognizer", img)
#         if detected_face:
#             break
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cam.release()
    cv2.destroyAllWindows()

# GUI setup
window = Tk()
window.title("Face Recognition Attendance System")

lbl1 = Label(window, text="Enter ID:")
lbl1.grid(row=0, column=0, padx=10, pady=10)
txt1 = Entry(window)
txt1.grid(row=0, column=1, padx=10, pady=10)

lbl2 = Label(window, text="Enter Name:")
lbl2.grid(row=1, column=0, padx=10, pady=10)
txt2 = Entry(window)
txt2.grid(row=1, column=1, padx=10, pady=10)

btn1 = Button(window, text="Take Images", command=take_images)
btn1.grid(row=2, column=0, padx=10, pady=10)

btn2 = Button(window, text="Train Images", command=train_images)
btn2.grid(row=2, column=1, padx=10, pady=10)

btn3 = Button(window, text="Recognize & Mark Attendance", command=recognize_faces)
btn3.grid(row=3, column=0, columnspan=2, padx=10, pady=10) 

window.mainloop()
```

```python
import cv2
import numpy as np
import os
from tkinter import *
from tkinter import messagebox, ttk
import pandas as pd
from datetime import datetime
import time

# Ensure folders exist
if not os.path.exists("TrainingImage"):
    os.makedirs("TrainingImage")
if not os.path.exists("TrainingImageLabel"):
    os.makedirs("TrainingImageLabel")
if not os.path.exists("Attendance"):
    os.makedirs("Attendance")

haarcascadePath = "haarcascade_frontalface_default.xml"

# ----------------- Backend Functions -----------------

def take_images(Id, name, notify):
    import time
    if not Id or not name:
        notify.set("Please fill both Enrollment No. and Name")
        return

    cam = cv2.VideoCapture(0)
    face_detector = cv2.CascadeClassifier(haarcascadePath)

    if not os.path.exists("TrainingImage"):
        os.makedirs("TrainingImage")

    # Poses in sequence
    pose_sequence = ["front", "left", "right"]
    pose_index = 0

    stable_frames_required = 8  # number of consecutive good frames required
    stable_frames_count = 0

    notify.set(f"Look {pose_sequence[pose_index].upper()} at the camera")

    def is_good_image(face_img):
        blur_score = cv2.Laplacian(face_img, cv2.CV_64F).var()
        if blur_score < 50:
            return False, "Image is blurry. Hold still."
        brightness = np.mean(face_img)
        if brightness < 60:
            return False, "Improve lighting. Too dark."
        return True, "Good image"

    while pose_index < len(pose_sequence):
        ret, img = cam.read()
        if not ret:
            notify.set("Camera not detected.")
            break

        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_detector.detectMultiScale(gray, 1.2, 5)

        if len(faces) == 0:
            notify.set("No face detected. Please adjust your face.")
            stable_frames_count = 0
        else:
            for (x, y, w, h) in faces:
                face_img = gray[y:y + h, x:x + w]
                good, msg = is_good_image(face_img)
                if good:
                    stable_frames_count += 1
                    notify.set(
                        f"{pose_sequence[pose_index].capitalize()} image detected ({stable_frames_count}/{stable_frames_required}). Hold still..."
                    )
                    # Draw rectangle for feedback
                    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
                    cv2.putText(
                        img,
                        f"{pose_sequence[pose_index]}",
                        (x, y - 10),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.8,
                        (255, 0, 0),
                        2,
                    )
                    if stable_frames_count >= stable_frames_required:
                        filename = f"TrainingImage/{name}.{Id}.{pose_sequence[pose_index]}.jpg"
                        cv2.imwrite(filename, face_img)
                        notify.set(
                            f"{pose_sequence[pose_index].capitalize()} image captured successfully!"
                        )
                        pose_index += 1
                        stable_frames_count = 0
                        if pose_index < len(pose_sequence):
                            notify.set(f"Now look {pose_sequence[pose_index].upper()}")
                        time.sleep(1)  # Wait to let user move to new pose
                else:
                    notify.set(msg)
                    stable_frames_count = 0
                # Draw rectangle for feedback even if image is not good
                cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
                cv2.putText(
                    img,
                    f"{pose_sequence[pose_index]}",
                    (x, y - 10),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.8,
                    (255, 0, 0),
                    2,
                )
        cv2.imshow("Register Face", img)
        k = cv2.waitKey(1)
        if k & 0xFF == ord("q"):
            notify.set("Image capture aborted.")
            break

    cam.release()
    cv2.destroyAllWindows()
    if pose_index == len(pose_sequence):
        notify.set("Face registration completed successfully!")

def train_images(notify):
    recognizer = cv2.face.LBPHFaceRecognizer_create()

    def getImagesAndLabels(path):
        imagePaths = [os.path.join(path, f) for f in os.listdir(path)]
        faces, Ids = [], []
        for imagePath in imagePaths:
            img = cv2.imread(imagePath, cv2.IMREAD_GRAYSCALE)
            if img is None:
                continue
            Id = int(os.path.split(imagePath)[-1].split(".")[1])
            faces.append(img)
            Ids.append(Id)
        return faces, Ids

    faces, Ids = getImagesAndLabels("TrainingImage")
    if len(faces) == 0:
        notify.set("No images found to train.")
        return
    recognizer.train(faces, np.array(Ids))
    recognizer.save("TrainingImageLabel/Trainer.yml")
    notify.set("Training completed!")

def mark_attendance(Id, name, subject):
    date_str = datetime.now().strftime("%Y-%m-%d")
    time_str = datetime.now().strftime("%H:%M:%S")
    filename = f"Attendance/{subject}_Attendance.csv"
    if not os.path.exists(filename):
        df = pd.DataFrame(columns=["Id", "Name", "Date", "Time"])
    else:
        df = pd.read_csv(filename)

    if not ((df["Id"] == Id) & (df["Date"] == date_str)).any():
        df = pd.concat([df, pd.DataFrame([{"Id": Id, "Name": name, "Date": date_str, "Time": time_str}])], ignore_index=True)
        df.to_csv(filename, index=False)

def recognize_faces(subject, notify):
    recognizer = cv2.face.LBPHFaceRecognizer_create()
    if not os.path.exists("TrainingImageLabel/Trainer.yml"):
        notify.set("Trainer not found.")
        return
    recognizer.read("TrainingImageLabel/Trainer.yml")

    faceCascade = cv2.CascadeClassifier(haarcascadePath)
    if faceCascade.empty():
        notify.set("Cascade not loaded!")
        return

    font = cv2.FONT_HERSHEY_SIMPLEX
    cam = cv2.VideoCapture(0)
    while True:
        ret, img = cam.read()
        if not ret:
            break
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = faceCascade.detectMultiScale(gray, 1.2, 5)
        for (x, y, w, h) in faces:
            Id, conf = recognizer.predict(gray[y:y+h, x:x+w])
            if conf < 60:
                matching_files = [f for f in os.listdir("TrainingImage") if f.split(".")[1] == str(Id)]
                name = matching_files[0].split(".")[0] if matching_files else "Unknown"
                mark_attendance(Id, name, subject)
                color, label = (0, 255, 0), name
            else:
                color, label = (0, 0, 255), "Unknown"
            cv2.putText(img, f"{label}", (x, y-10), font, 0.8, color, 2)
            cv2.rectangle(img, (x, y), (x+w, y+h), color, 3)
        cv2.imshow("Face Recognizer", img)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cam.release()
    cv2.destroyAllWindows()
    notify.set(f"Attendance marked for {subject}")

# ----------------- UI Functions -----------------

# ----------------- Theme Colors -----------------
BACKGROUND = "#e6f2ff"   # very light blue
FRAME_BG   = "#cce6ff"   # light blue
BUTTON_BG  = "#3399ff"   # bright blue
BUTTON_FG  = "white"
TEXT_FG    = "black"
TITLE_FG   = "#003366"   # dark navy

# ----------------- UI Functions -----------------

def open_register_window():
    reg_win = Toplevel(window)
    reg_win.title("Register Student")
    reg_win.configure(bg=BACKGROUND)

    Label(reg_win, text="Register Your Face", font=("Arial", 20, "bold"), 
          fg=TITLE_FG, bg=BACKGROUND).pack(pady=10)
    Label(reg_win, text="Enter the details", font=("Arial", 16, "bold"), 
          fg=TEXT_FG, bg=BACKGROUND).pack(pady=5)

    frm = Frame(reg_win, bg=FRAME_BG, padx=20, pady=20)
    frm.pack(pady=10)

    Label(frm, text="Enrollment No:", fg=TEXT_FG, bg=FRAME_BG, font=("Arial", 12)).grid(row=0, column=0, sticky=W, pady=5)
    id_entry = Entry(frm, font=("Arial", 12))
    id_entry.grid(row=0, column=1, pady=5)

    Label(frm, text="Name:", fg=TEXT_FG, bg=FRAME_BG, font=("Arial", 12)).grid(row=1, column=0, sticky=W, pady=5)
    name_entry = Entry(frm, font=("Arial", 12))
    name_entry.grid(row=1, column=1, pady=5)

    notify = StringVar()
    Label(frm, textvariable=notify, fg="red", bg=FRAME_BG, font=("Arial", 12)).grid(row=2, column=0, columnspan=2, pady=10)

    Button(frm, text="Take Image", bg=BUTTON_BG, fg=BUTTON_FG, font=("Arial", 12, "bold"), 
           command=lambda: take_images(id_entry.get(), name_entry.get(), notify)).grid(row=3, column=0, padx=10, pady=10)
    Button(frm, text="Train Image", bg=BUTTON_BG, fg=BUTTON_FG, font=("Arial", 12, "bold"), 
           command=lambda: train_images(notify)).grid(row=3, column=1, padx=10, pady=10)


def open_subject_window():
    sub_win = Toplevel(window)
    sub_win.title("Subject")
    sub_win.configure(bg=BACKGROUND)

    Label(sub_win, text="Which Subject of Attendance?", font=("Arial", 16, "bold"), 
          fg=TITLE_FG, bg=BACKGROUND).pack(pady=10)

    frm = Frame(sub_win, bg=FRAME_BG, padx=20, pady=20)
    frm.pack(pady=10)

    Label(frm, text="Enter Subject:", fg=TEXT_FG, bg=FRAME_BG, font=("Arial", 12)).grid(row=0, column=0)
    sub_entry = Entry(frm, font=("Arial", 12))
    sub_entry.grid(row=0, column=1, padx=10)

    notify = StringVar()
    Label(frm, textvariable=notify, fg="red", bg=FRAME_BG, font=("Arial", 12)).grid(row=1, column=0, columnspan=2, pady=10)

    Button(frm, text="Take Attendance", bg=BUTTON_BG, fg=BUTTON_FG, font=("Arial", 12, "bold"), 
           command=lambda: recognize_faces(sub_entry.get(), notify)).grid(row=2, column=0, pady=10)
    Button(frm, text="View Attendance", bg=BUTTON_BG, fg=BUTTON_FG, font=("Arial", 12, "bold"), 
           command=lambda: view_attendance(sub_entry.get())).grid(row=2, column=1, pady=10)


def view_attendance(subject):
    filename = f"Attendance/{subject}_Attendance.csv"
    if not os.path.exists(filename):
        messagebox.showerror("Error", f"No attendance found for {subject}")
        return

    df = pd.read_csv(filename)
    view_win = Toplevel(window)
    view_win.title(f"Attendance of {subject}")
    view_win.configure(bg=BACKGROUND)

    cols = list(df.columns)
    tree = ttk.Treeview(view_win, columns=cols, show="headings")
    style = ttk.Style()
    style.configure("Treeview", background=FRAME_BG, foreground=TEXT_FG, 
                    fieldbackground=FRAME_BG, rowheight=30, font=("Arial", 12))
    style.configure("Treeview.Heading", font=("Arial", 12, "bold"), foreground=TITLE_FG)

    for col in cols:
        tree.heading(col, text=col)
        tree.column(col, width=150)
    for _, row in df.iterrows():
        tree.insert("", "end", values=list(row))
    tree.pack(fill=BOTH, expand=True)

# ----------------- Main UI -----------------

window = Tk()
window.title("CLASS VISION")
window.configure(bg=BACKGROUND)
window.geometry("1000x700")

Label(window, text="CLASS VISION", font=("Arial", 22, "bold"), fg=TITLE_FG, bg=BACKGROUND).pack(pady=10)
Label(window, text="Welcome to CLASS VISION", font=("Arial", 20, "bold"), fg=TEXT_FG, bg=BACKGROUND).pack(pady=10)

frame = Frame(window, bg=FRAME_BG, padx=40, pady=40)
frame.pack(pady=60)

Button(frame, text="Register a new student", command=open_register_window, font=("Arial", 14, "bold"),
       bg=BUTTON_BG, fg=BUTTON_FG, width=20, height=2).grid(row=0, column=0, padx=40)
Button(frame, text="Take Attendance", command=open_subject_window, font=("Arial", 14, "bold"),
       bg=BUTTON_BG, fg=BUTTON_FG, width=20, height=2).grid(row=0, column=1, padx=40)

Button(window, text="EXIT", command=window.destroy, font=("Arial", 14, "bold"),
       bg="red", fg="white", width=15, height=2).pack(pady=40)

window.mainloop()
```

```python
import cv2
import numpy as np
import os
from tkinter import *
from tkinter import messagebox, ttk
import pandas as pd
from datetime import datetime
import time
import warnings
warnings.filterwarnings("ignore")

# ---------- Ensure folders ----------
if not os.path.exists("TrainingImage"):
    os.makedirs("TrainingImage")
if not os.path.exists("TrainingImageLabel"):
    os.makedirs("TrainingImageLabel")
if not os.path.exists("Attendance"):
    os.makedirs("Attendance")

haarcascadePath = "haarcascade_frontalface_default.xml"

# ---------- Backend Functions ----------

def take_images(Id, name, notify):
    """
    Face registration:
    - Steps through a pose sequence (front/left/right/up/down)
    - Requires a number of stable good frames per pose (blur + brightness checks)
    - Saves one image per pose after the stable window
    - Provides live feedback via `notify` StringVar and forces Tkinter update
    """
    if not Id or not name:
        notify.set("Please fill both Enrollment No. and Name")
        try:
            window.update()
        except:
            pass
        return

    # Load detector
    face_detector = cv2.CascadeClassifier(haarcascadePath)
    if face_detector.empty():
        notify.set("Failed to load cascade. Check haarcascade file path.")
        try:
            window.update()
        except:
            pass
        return

    cam = cv2.VideoCapture(0)
    if not cam.isOpened():
        notify.set("Cannot open camera.")
        try:
            window.update()
        except:
            pass
        return

    # Pose sequence (can add/remove)
    pose_sequence = ["front", "left", "right", "up", "down"]
    pose_index = 0
    stable_frames_required = 8   # frames required while face is steady and good
    stable_frames_count = 0

    notify.set(f"Please look {pose_sequence[pose_index].upper()}")
    try:
        window.update()
    except:
        pass

    def is_good_image(face_img):
        """
        Return (True/False, message) based on blur and brightness tests.
        Adjust thresholds as needed for your environment.
        """
        # Blur check (Laplacian variance)
        blur_score = cv2.Laplacian(face_img, cv2.CV_64F).var()
        if blur_score < 50:
            return False, "Image is blurry. Hold still."
        # Brightness check
        brightness = np.mean(face_img)
        if brightness < 60:
            return False, "Too dark. Increase lighting."
        if brightness > 220:
            return False, "Too bright / overexposed. Adjust lighting."
        return True, "Good image"

    saved_count = 0
    last_notify_time = 0

    try:
        while pose_index < len(pose_sequence):
            ret, img = cam.read()
            if not ret:
                notify.set("Camera frame not received.")
                window.update()
                break

            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = face_detector.detectMultiScale(gray, 1.2, 5)

            if len(faces) == 0:
                # No faces found  prompt user
                notify.set(f"No face detected. Please move face into view.")
                stable_frames_count = 0
                # update UI occasionally to avoid too many updates
                if time.time() - last_notify_time > 0.2:
                    try:
                        window.update()
                    except:
                        pass
                    last_notify_time = time.time()
            else:
                # Use the largest detected face (likely the user)
                faces = sorted(faces, key=lambda r: r[2]*r[3], reverse=True)
                (x, y, w, h) = faces[0]
                face_img = gray[y:y+h, x:x+w]

                good, msg = is_good_image(face_img)
                if good:
                    stable_frames_count += 1
                    notify.set(
                        f"{pose_sequence[pose_index].capitalize()} detected ({stable_frames_count}/{stable_frames_required})  Hold still..."
                    )
                    # update UI
                    try:
                        window.update()
                    except:
                        pass

                    # Draw green box & label
                    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
                    if pose_index < len(pose_sequence):
                        cv2.putText(img, pose_sequence[pose_index].upper(), (x, y - 10),
                                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

                    # When stable window reached, save one image and advance
                    if stable_frames_count >= stable_frames_required:
                        filename = f"TrainingImage/{name}.{Id}.{pose_sequence[pose_index]}.jpg"
                        # Save resized face for consistency (optional)
                        try:
                            saved_face = cv2.resize(face_img, (200, 200))
                        except:
                            saved_face = face_img
                        cv2.imwrite(filename, saved_face)
                        saved_count += 1
                        notify.set(f"{pose_sequence[pose_index].capitalize()} captured.")
                        try:
                            window.update()
                        except:
                            pass
                        pose_index += 1
                        stable_frames_count = 0
                        # small pause so user can read the message
                        time.sleep(0.9)
                        if pose_index < len(pose_sequence):
                            notify.set(f"Now look {pose_sequence[pose_index].upper()}")
                            try:
                                window.update()
                            except:
                                pass
                else:
                    # Bad quality: reset stable counter, show message
                    notify.set(msg)
                    stable_frames_count = 0
                    try:
                        window.update()
                    except:
                        pass
                    # Draw red box & feedback
                    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 2)
                    if pose_index < len(pose_sequence):
                        cv2.putText(img, pose_sequence[pose_index].upper(), (x, y - 10),
                                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

            # Show frame
            cv2.imshow("Register Face", img)
            k = cv2.waitKey(1) & 0xFF
            if k == ord("q"):
                notify.set("Image capture aborted by user.")
                try:
                    window.update()
                except:
                    pass
                break

    finally:
        cam.release()
        cv2.destroyAllWindows()

    if pose_index >= len(pose_sequence):
        notify.set("Face registration completed successfully!")
    else:
        # If aborted or camera failed
        if saved_count > 0:
            notify.set(f"Partial registration done ({saved_count} poses). Re-run to complete.")
        else:
            notify.set("No images captured. Please try again.")
    try:
        window.update()
    except:
        pass


def train_images(notify):
    """Train LBPH recognizer using images in TrainingImage/"""
    # Ensure opencv-contrib is installed for cv2.face
    try:
        recognizer = cv2.face.LBPHFaceRecognizer_create()
    except Exception as e:
        notify.set("OpenCV face module not available. Install opencv-contrib.")
        try: window.update()
        except: pass
        return

    def getImagesAndLabels(path):
        imagePaths = [os.path.join(path, f) for f in os.listdir(path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        faces, Ids = [], []
        for imagePath in imagePaths:
            img = cv2.imread(imagePath, cv2.IMREAD_GRAYSCALE)
            if img is None:
                continue
            fname = os.path.split(imagePath)[-1]
            parts = fname.split(".")
            # Expected filename format: name.Id.pose.jpg OR name.Id.pose.N.jpg
            # Find the numeric Id anywhere in parts
            Id = None
            for p in parts:
                if p.isdigit():
                    Id = int(p)
                    break
            if Id is None:
                continue
            faces.append(img)
            Ids.append(Id)
        return faces, Ids

    faces, Ids = getImagesAndLabels("TrainingImage")
    if len(faces) == 0:
        notify.set("No images found to train.")
        try: window.update()
        except: pass
        return

    recognizer.train(faces, np.array(Ids))
    recognizer.save("TrainingImageLabel/Trainer.yml")
    notify.set("Training completed!")
    try: window.update()
    except: pass


def mark_attendance(Id, name, subject="General"):
    """Mark attendance with Id, Name, Date, Time. Fix CSV headers if needed."""
    date_str = datetime.now().strftime("%Y-%m-%d")
    time_str = datetime.now().strftime("%H:%M:%S")
    filename = f"Attendance/{subject}_Attendance.csv"
    expected_cols = ["Id", "Name", "Date", "Time"]

    if not os.path.exists(filename):
        df = pd.DataFrame(columns=expected_cols)
    else:
        df = pd.read_csv(filename)
        # If columns are missing or wrong, reset to expected
        if not set(expected_cols).issubset(df.columns):
            df = pd.DataFrame(columns=expected_cols)

    # Avoid duplicate for same day
    if not ((df["Id"] == Id) & (df["Date"] == date_str)).any():
        new_row = {"Id": Id, "Name": name, "Date": date_str, "Time": time_str}
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
        df.to_csv(filename, index=False)


def recognize_faces(subject, notify):
    """Run live recognition and mark attendance for recognized faces"""
    if not os.path.exists("TrainingImageLabel/Trainer.yml"):
        notify.set("Trainer not found. Train first.")
        try: window.update()
        except: pass
        return

    try:
        recognizer = cv2.face.LBPHFaceRecognizer_create()
    except Exception:
        notify.set("OpenCV face module not available. Install opencv-contrib.")
        try: window.update()
        except: pass
        return

    recognizer.read("TrainingImageLabel/Trainer.yml")
    faceCascade = cv2.CascadeClassifier(haarcascadePath)
    if faceCascade.empty():
        notify.set("Cascade not loaded. Check Haarcascade file.")
        try: window.update()
        except: pass
        return

    cam = cv2.VideoCapture(0)
    if not cam.isOpened():
        notify.set("Cannot open camera.")
        try: window.update()
        except: pass
        return

    font = cv2.FONT_HERSHEY_SIMPLEX
    notify.set("Starting recognition... Press 'q' to stop.")
    try: window.update()
    except: pass

    #  Track marked IDs this session
    marked_today = set()

    while True:
        ret, img = cam.read()
        if not ret:
            break
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = faceCascade.detectMultiScale(gray, 1.2, 5)

        for (x, y, w, h) in faces:
            face_region = gray[y:y+h, x:x+w]
            try:
                Id, conf = recognizer.predict(face_region)
            except Exception:
                Id, conf = -1, 1000

            if conf < 70:   # threshold
                # find name for the Id
                matching_files = [f for f in os.listdir("TrainingImage")
                                  if len(f.split("."))>1 and f.split(".")[1].isdigit()
                                  and int(f.split(".")[1])==int(Id)]
                name = matching_files[0].split(".")[0] if matching_files else str(Id)

                #  Only mark once per session
                if Id not in marked_today:
                    mark_attendance(str(Id), name, subject)
                    marked_today.add(Id)
                    notify.set(f"Attendance marked for {name}")
                else:
                    notify.set(f"{name} already marked this session")

                color, label = (0, 255, 0), f"{name}"
            else:
                color, label = (0, 0, 255), "Unknown"
                notify.set("Unknown face detected")

            try: window.update()
            except: pass

            cv2.putText(img, f"{label}", (x, y-10), font, 0.8, color, 2)
            cv2.rectangle(img, (x, y), (x+w, y+h), color, 3)

        cv2.imshow("Face Recognizer", img)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cam.release()
    cv2.destroyAllWindows()
    notify.set(f"Recognition stopped for {subject}")
    try: window.update()
    except: pass


# ---------- UI Functions & Theme ----------
BACKGROUND = "#e6f2ff"   # very light blue
FRAME_BG   = "#cce6ff"   # light blue
BUTTON_BG  = "#3399ff"   # bright blue
BUTTON_FG  = "white"
TEXT_FG    = "black"
TITLE_FG   = "#003366"   # dark navy

def open_register_window():
    reg_win = Toplevel(window)
    reg_win.title("Register Student")
    reg_win.configure(bg=BACKGROUND)

    Label(reg_win, text="Register Your Face", font=("Arial", 20, "bold"),
          fg=TITLE_FG, bg=BACKGROUND).pack(pady=10)
    Label(reg_win, text="Enter the details", font=("Arial", 16, "bold"),
          fg=TEXT_FG, bg=BACKGROUND).pack(pady=5)

    frm = Frame(reg_win, bg=FRAME_BG, padx=20, pady=20)
    frm.pack(pady=10)

    Label(frm, text="Enrollment No:", fg=TEXT_FG, bg=FRAME_BG, font=("Arial", 12)).grid(row=0, column=0, sticky=W, pady=5)
    id_entry = Entry(frm, font=("Arial", 12))
    id_entry.grid(row=0, column=1, pady=5)

    Label(frm, text="Name:", fg=TEXT_FG, bg=FRAME_BG, font=("Arial", 12)).grid(row=1, column=0, sticky=W, pady=5)
    name_entry = Entry(frm, font=("Arial", 12))
    name_entry.grid(row=1, column=1, pady=5)

    notify = StringVar()
    Label(frm, textvariable=notify, fg="red", bg=FRAME_BG, font=("Arial", 12)).grid(row=2, column=0, columnspan=2, pady=10)

    Button(frm, text="Take Image", bg=BUTTON_BG, fg=BUTTON_FG, font=("Arial", 12, "bold"),
           command=lambda: take_images(id_entry.get(), name_entry.get(), notify)).grid(row=3, column=0, padx=10, pady=10)
    Button(frm, text="Train Image", bg=BUTTON_BG, fg=BUTTON_FG, font=("Arial", 12, "bold"),
           command=lambda: train_images(notify)).grid(row=3, column=1, padx=10, pady=10)


def open_subject_window():
    sub_win = Toplevel(window)
    sub_win.title("Subject")
    sub_win.configure(bg=BACKGROUND)

    Label(sub_win, text="Which Subject of Attendance?", font=("Arial", 16, "bold"),
          fg=TITLE_FG, bg=BACKGROUND).pack(pady=10)

    frm = Frame(sub_win, bg=FRAME_BG, padx=20, pady=20)
    frm.pack(pady=10)

    Label(frm, text="Enter Subject:", fg=TEXT_FG, bg=FRAME_BG, font=("Arial", 12)).grid(row=0, column=0)
    sub_entry = Entry(frm, font=("Arial", 12))
    sub_entry.grid(row=0, column=1, padx=10)

    notify = StringVar()
    Label(frm, textvariable=notify, fg="red", bg=FRAME_BG, font=("Arial", 12)).grid(row=1, column=0, columnspan=2, pady=10)

    Button(frm, text="Take Attendance", bg=BUTTON_BG, fg=BUTTON_FG, font=("Arial", 12, "bold"),
           command=lambda: recognize_faces(sub_entry.get() or "General", notify)).grid(row=2, column=0, pady=10)
    Button(frm, text="View Attendance", bg=BUTTON_BG, fg=BUTTON_FG, font=("Arial", 12, "bold"),
           command=lambda: view_attendance(sub_entry.get() or "General")).grid(row=2, column=1, pady=10)


def view_attendance(subject):
    filename = f"Attendance/{subject}_Attendance.csv"
    if not os.path.exists(filename):
        messagebox.showerror("Error", f"No attendance found for {subject}")
        return

    df = pd.read_csv(filename)
    view_win = Toplevel(window)
    view_win.title(f"Attendance of {subject}")
    view_win.configure(bg=BACKGROUND)

    cols = list(df.columns)
    tree = ttk.Treeview(view_win, columns=cols, show="headings")
    style = ttk.Style()
    style.configure("Treeview", background=FRAME_BG, foreground=TEXT_FG,
                    fieldbackground=FRAME_BG, rowheight=30, font=("Arial", 12))
    style.configure("Treeview.Heading", font=("Arial", 12, "bold"), foreground=TITLE_FG)

    for col in cols:
        tree.heading(col, text=col)
        tree.column(col, width=150)
    for _, row in df.iterrows():
        tree.insert("", "end", values=list(row))
    tree.pack(fill=BOTH, expand=True)


# ---------- Main UI ----------
window = Tk()
window.title("CLASS VISION")
window.configure(bg=BACKGROUND)
window.geometry("1000x700")

Label(window, text="CLASS VISION", font=("Arial", 22, "bold"), fg=TITLE_FG, bg=BACKGROUND).pack(pady=10)
Label(window, text="Welcome to CLASS VISION", font=("Arial", 20, "bold"), fg=TEXT_FG, bg=BACKGROUND).pack(pady=10)

frame = Frame(window, bg=FRAME_BG, padx=40, pady=40)
frame.pack(pady=60)

Button(frame, text="Register a new student", command=open_register_window, font=("Arial", 14, "bold"),
       bg=BUTTON_BG, fg=BUTTON_FG, width=20, height=2).grid(row=0, column=0, padx=40)
Button(frame, text="Take Attendance", command=open_subject_window, font=("Arial", 14, "bold"),
       bg=BUTTON_BG, fg=BUTTON_FG, width=20, height=2).grid(row=0, column=1, padx=40)

Button(window, text="EXIT", command=window.destroy, font=("Arial", 14, "bold"),
       bg="red", fg="white", width=15, height=2).pack(pady=40)

window.mainloop()
```
